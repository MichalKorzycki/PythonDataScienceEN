```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("dark_background")

x = np.linspace(-1, 1, 100)
fig = plt.figure(figsize=(10,10))

k=2
plt.plot(x, 1/(1+np.exp(-k*x)))

k=4
plt.plot(x, 1/(1+np.exp(-k*x)))

k=6
plt.plot(x, 1/(1+np.exp(-k*x)))

k=8
plt.plot(x, 1/(1+np.exp(-k*x)))
```
